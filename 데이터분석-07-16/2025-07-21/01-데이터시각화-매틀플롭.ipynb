```python
# Matplotlib (매트플롯 라이브러리)
# 그래프를 그릴 수 있는 파이썬 라이브러리
import matplotlib.pyplot as plt

x = [i for i in range(1, 6)] # comprehension
y = [i for i in range(1, 6)]

plt.plot(x,y)
plt.title("첫번째 그래프")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
# 객체 기반 스타일로 그래프 그리기
x = [i for i in range(1, 6)] # comprehension
y = [i for i in range(1, 6)]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("x")
ax.set_ylabel("y")

fig.set_dpi(300)
fig.savefig("first_plot.png")
```

```python
# 객체 기반 스타일로 그래프 그리기
# 하나의 도화지에 그래프 두 개 그려보기
# fig는 전체 도화지를 의미함
# ax는 그래프를 그릴 수 있는곳
# 여러 개의 그래프, 세밀한 제어가 필요할 때

#  1. Figure와 2개의 Axes 생성 (1행 2열 구조)
fig, ax = plt.subplots(1, 2, figsize=(10, 4)) # 가로 10인치 세로 4인치

# 2. 첫 번째 그래프 (왼쪽)
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
ax[0].plot(x, y1, color='blue', marker='o')
ax[0].set_title("first 1")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

# 2. 두 번째 그래프 (오른쪽)
x = [1, 2, 3, 4]
y2 = [30, 25, 20, 10]
ax[1].plot(x, y2, color='green', marker='x')
ax[1].set_title("second 2")
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")
```

```python
# 2. (pyplot 스타일)
# plt 1개짜리 간단하고 빠르게 그릴 때 
# plt 객체에 있는 함수를 바로 이용

import matplotlib.pyplot as plt

# 그래프 그리기
plt.plot([1, 2, 3], [1, 4, 9])

# 제목과 레이블
plt.title("pyplot style")
plt.xlabel("x")
plt.ylabel("y")

# 그래프 보여주기
plt.show()
```

```python
# 1.(Line plot) 선 그래프
# 용도 :  선그래프에서는 범주형 데이터(연속형)를 x축에, 연속형 데이터를 y축에 사용하는 것이 일반적임
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']  # 범주형
study_hours = [2.5, 3.0, 4.0, 3.5, 2.0, 1.0, 0.5]         # 연속형

plt.plot(days, study_hours, marker='o', linestyle='-', color='blue')
plt.title('Average Study Time by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Study Time (hours)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0., 5., 0.2)
# 빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.plot(t, t, 'r--', # 빨간색(red) 점선
         t, t**2, 'bs', # 파란색(blue) 사각형(square)
         t, t**3, 'g^') #녹색(green) 삼각형(위쪽을 향함)
#plt.show()    .py 에서는 이걸 해줘야 그래프가 뜬다. 
```

```python
#### 눈금선 표시
x = np.arange(1, 11)
y1 = x ** 2
y2 = x ** 3
y3 = x * 4
plt.plot(x, y1,'ro--', x, y2, 'gs--', x, y3, 'b^--')
plt.legend(['red', 'green', 'blue'], loc='best')
# 눈금선이 표시되도록 설정
# plt.grid(True)
# 눈금선의 스타일 설정
# describes a sequence of 5 point and 1 point dashes separated by 2 
plt.grid(which='both', axis='x', color='green', alpha=0.5, dashes=(5,2,1,2)) # which 옵션은 major 나 minor 로 설정될 수 있음
plt.grid(which='both', axis='y', color='red', alpha=0.5, linestyle='-.', linewidth=1)
# plt.grid(False) 그리드를 사라지게 함
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
# 데이터
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
ax.plot(x, y)
# 보조 눈금 활성화
ax.minorticks_on()
# 주 눈금선만
ax.grid(which='major', linestyle='-', color='blue')
# 보조 눈금선만
ax.grid(which='minor', linestyle=':', color='orange')
plt.show()
```

```python
# subplot 그래프 여러개 그리기
name = ["Kim", "Yoon", "Choi", "Park"]
month = ["March", "April", "May", "June"]
data = {'name'    : month,
        'english' : [80, 90, 100, 95],
        'chinese' : [100, 80, 70, 85],
        'korean'  : [95, 100, 80, 60],
        'math'    : [90, 70, 60, 100]}
df = pd.DataFrame(data)
df
```

```python
plt.subplot(2,2,1) # 배치 2행 2열 첫번재

plt.plot('name', 'english', 'ro--', data=df)
plt.ylim(60, 100)
plt.grid()
plt.subplot(2,2,2) # 배치 2행 2열 두번째
plt.plot('name', 'chinese', 'gs--', data=df)
plt.ylim(60, 100)
plt.grid()
plt.subplot(2,2,3)
plt.plot('name', 'korean', 'b^--', data=df)
plt.ylim(60, 100)
plt.grid()
plt.subplot(224) # 하나의 숫자로 표현
plt.plot('name', 'math', 'm--', data=df)
plt.ylim(60, 100)
plt.grid()
plt.show()
```

```python
# Bar 차트 
# Bar 그래프(범주형 데이터를 시각화하는 데 가장 적합)
# x 축이 순서 없는 범주형
# 누가 더 많고 적은지 쉽게 파악 가능
# Bar 차트는 "무엇이 얼마나 많은가?" 를 비교할 때 가장 직관적

names = ['A', 'B', 'O']
values = [50, 30, 100]

plt.subplot(121)
plt.bar(names, values)

plt.subplot(122)
plt.barh(names, values) # 가로 막대 그래프
plt.show()
```

```python
# 히스토그램
# 수치형(연속형) 데이터에 적합
# 히스토그램으로 얻을 수 있는 통찰
# 데이터가 정규분포를 따르는지 체크 가능
# 편향(skewed) 되어 있는지 (왼쪽/오른쪽) 체크
# 이상치가 존재하는지 체크 유무

np.random.seed(1)

# 평균이 100, 표춘편차가 15인 정규분포 데이터를 10,000개 생성. 
x = np.random.normal(100, 15, 10000)

# 전체 범위에 대해(별도 지정 필요 없음), 50개 계급으로 나눈 히스토그램 그리기
plt.hist(x, bins=50)
plt.grid()
plt.show()
```

```python
# 학생 100명의 시험 점수(0~100 사이) 를 무작위로 생성한 뒤,
# 그 분포를 히스토그램으로 시각화
import numpy as np
import matplotlib.pyplot as plt
# 1. 데이터 생성 (0~100 사이의 점수, 정규분포 기반)
np.random.seed(42)  # 결과 고정
scores = np.random.normal(loc=70, scale=10, size=100)  # 평균 70, 표준편차 10
print(scores)

# 2. 히스토그램 그리기
# 점수를 10개의 구간으로 나눔 (예: 5060, 6070 등)
plt.hist(scores, bins=10, color='skyblue', edgecolor='black')  # 10개 구간으로 나눔
plt.title('학생 시험 점수 분포')
plt.xlabel('점수 구간')
plt.ylabel('학생 수')
plt.grid(True)
# 3. 그래프 출력
plt.show()
```

```python
# 파이(Pie) 그래프
# 범주형 데이터 (Categorical Data)
# 과일 종류별 판매량
# 학년별 평균 점수
# 도시별 인구
# 언어별 선호도

# 이산형 수치 데이터 (Discrete Numerical Data)
# 1부터 5까지의 만족도 점수별 인원 수
# A~F 학점 에 대한 인원수

import matplotlib.pyplot as plt

color = ['yellowgreen', 'gold', 'deepskyblue', 'red', 'lightgrey']
langs = ['Ruby', 'C', 'Java', 'Python', 'HTML5']
students = [7, 20, 15, 40, 10]

#가로 6인치, 세로 4인치 크기의 그래프 영역(Figure)을 생성
plt.figure(figsize=(6,4))

#autopct='%.2f%%': 각 조각에 비율을 소수점 둘째 자리까지 표시
# 40 / 총합 × 100
plt.pie(students, labels=langs, colors = color, autopct='%.2f%%')

#plt.axis('equal') #정확한 원형(circle) 유지
plt.title('Pie Chart')

plt.legend(loc='lower right',
           bbox_to_anchor=(1, 0), bbox_transform=plt.gcf().transFigure)

plt.show()
```
